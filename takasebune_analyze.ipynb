テキストファイルからの読み込みを行う
その結果を出力する

```python
import re
with open("takasebune.txt", mode="r") as f:
    takasebune_original = f.read()
print(takasebune_original)
```

前処理を行う。
本文以外の削除、ルビの削除、改行そのほかを削除

```python
takasebune=re.split(r'-------------------------------------------------------',takasebune_original)[2]
takasebune=re.split(r'底本：',takasebune)[0]
takasebune=re.sub("《[^》]+》", "", takasebune)
takasebune = re.sub("［[^］]+］", "", takasebune)
takasebune = re.sub("[｜ 　「」\n]", "", takasebune)
print(takasebune)
```

Pickleファイルに整形したデータを収納

```python
import pickle
with open("takasebune.pickle",mode="wb")as f:
    pickle.dump(takasebune,f)
```

MeCabをインポートし、Tagをつける

```python
%pip install mecab-python-windows
%pip install unidic-lite
import MeCab
with open("takasebune.pickle",mode="rb") as f:
    takasebune=pickle.load(f)
tag =MeCab.Tagger()
res =tag.parse(takasebune)
print(res)
```

名詞の選別ができているか出力で確認

```python
noun=[]
verb= []
adjective= []
adverb= []
postpos= []
auxiliaryverb= []

prs = tag.parseToNode(takasebune)
while prs:
    if prs.feature.split(',')[0] == "名詞":
        noun.append(prs.surface)
    elif prs.feature.split(',')[0] == "動詞":
        verb.append(prs.surface)
    elif prs.feature.split(',')[0] == "形容詞":
        adjective.append(prs.surface)
    elif prs.feature.split(',')[0] == "副詞":
        adverb.append(prs.surface)
    elif prs.feature.split(',')[0] == "助詞":
        postpos.append(prs.surface)
    elif prs.feature.split(',')[0] == "助動詞":
        auxiliaryverb.append(prs.surface)   
    prs = prs.next

print(noun)
```

品詞の数を出力

```python
print(len(noun))
print(len(verb))
print(len(adjective))
print(len(adverb))
print(len(postpos))
print(len(auxiliaryverb))
```

同様の処理をJanomeでも実施
まずは品詞ごとに分けられているかを確認

```python
%pip install janome
from janome.tokenizer import Tokenizer
t = Tokenizer()
for token in t.tokenize(takasebune):
    print(token)
```

名詞のみを抽出

```python
from janome.analyzer import Analyzer
from janome.tokenfilter import *
from janome.charfilter import *
partofspeech="名詞"
chrfilt = [UnicodeNormalizeCharFilter(), RegexReplaceCharFilter(r"[IiⅠｉ?.*/~=()〝 &lt;&gt;:：《°!！!？（）-]+", "")]
tknfilt = [POSKeepFilter([partofspeech]), POSStopFilter([]), LowerCaseFilter()]
analyzer = Analyzer(char_filters=chrfilt,tokenizer=t,token_filters=tknfilt)
noun= [token.surface for token in analyzer.analyze(takasebune)]

print(noun)
```

動詞以下、同様に処理を実施し、品詞の数をカウント

```python
print(len(noun))
partofspeech="動詞"
tknfilt = [POSKeepFilter([partofspeech]), POSStopFilter([]), LowerCaseFilter()]
analyzer = Analyzer(char_filters=chrfilt,tokenizer=t,token_filters=tknfilt)
verb= [token.surface for token in analyzer.analyze(takasebune)]
print(len(verb))
partofspeech="形容詞"
tknfilt = [POSKeepFilter([partofspeech]), POSStopFilter([]), LowerCaseFilter()]
analyzer = Analyzer(char_filters=chrfilt,tokenizer=t,token_filters=tknfilt)
adjective= [token.surface for token in analyzer.analyze(takasebune)]
print(len(adjective))
partofspeech="副詞"
tknfilt = [POSKeepFilter([partofspeech]), POSStopFilter([]), LowerCaseFilter()]
analyzer = Analyzer(char_filters=chrfilt,tokenizer=t,token_filters=tknfilt)
adverb= [token.surface for token in analyzer.analyze(takasebune)]
print(len(adverb))

partofspeech="助詞"
tknfilt = [POSKeepFilter([partofspeech]), POSStopFilter([]), LowerCaseFilter()]
analyzer = Analyzer(char_filters=chrfilt,tokenizer=t,token_filters=tknfilt)
postpos= [token.surface for token in analyzer.analyze(takasebune)]
print(len(postpos))

partofspeech="助動詞"
tknfilt = [POSKeepFilter([partofspeech]), POSStopFilter([]), LowerCaseFilter()]
analyzer = Analyzer(char_filters=chrfilt,tokenizer=t,token_filters=tknfilt)
auxiliaryverb= [token.surface for token in analyzer.analyze(takasebune)]
print(len(auxiliaryverb))
```
